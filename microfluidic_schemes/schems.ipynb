Membrane Chamber (I8)

```python
import schemdraw
import schemdraw.elements as elm
from schemdraw.segments import Segment, SegmentCircle, SegmentText

class MeterM(elm.Element2Term):
    _element_defaults = {
        'radius': 0.35,
        'text': 'M',
        'fontsize': 14,
    }

    def __init__(self, **kwargs):
        super().__init__(**kwargs)

        r = self.params['radius']
        t = self.params['text']
        fs = self.params['fontsize']

        cx, cy = 0.5, 0
        left_gap = cx - r
        right_gap = cx + r

        # ✅ 1) 隐形主路径：只用于 Element2Term 计算 istart/iend
        #    关键：放在 segments[0]，并且“不可见”
        self.segments.insert(0, Segment([(0, 0), (1, 0)], visible=False))

        # ✅ 2) 可见导线：两段线，中间留空，不横穿圆
        self.segments.append(Segment([(0, 0), (left_gap, 0)]))
        self.segments.append(Segment([(right_gap, 0), (1, 0)]))

        # ✅ 3) 圆与文字
        self.segments.append(SegmentCircle((cx, cy), r, fill=None))
        self.segments.append(
            SegmentText((cx, cy), t, align=('center', 'center'),
                        fontsize=fs, rotation_global=True)
        )

        self.anchors['center'] = (cx, cy)


with schemdraw.Drawing() as d:
    d += MeterM()
```
